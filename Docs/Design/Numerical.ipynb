```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.zeros((100, 100))

for i in range(1, 101):
    for j in range(1, 101):
        p = (i + j) / 2
        arr[i-1, j-1] = p
```

```python
im = plt.imshow(arr)
```

```python
flat_arr = arr.reshape(-1)
plist = np.zeros(100)

for i in range (1, 101):
    res = list(filter(lambda x: x <= i, flat_arr))
    plist[i-1] = len(res) / (100 * 100)

plt.plot(plist)
```

它似乎被定义为上是两个均匀分布的随机变量之和的联合分布函数（？）

闭包形式的解需要使用数学卷积的知识。

```python
for i in range(0, 100, 10):
    print(i, str(plist[i] * 100) + '%')
```
